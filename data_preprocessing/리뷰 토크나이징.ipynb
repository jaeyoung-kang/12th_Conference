```python
import re
```

```python
# 이모티콘 제거
def rm_emo(text):
    return re.sub('🐈|💛|✔️|🎃|🥰|🌮|☕️|🍹|🍞|🥯|🥐|🥖|🥨|🍓|👍|👏|😋|💕|😂|🍯|🐑|😗|📍|🧼|😭|티|🙆|❤️|프🏻🏻|🍪|💘|이🏻🏻🏻|!🏻|😊|🙅🏻|🤩|🦁|음🏻🏻|👉|요🏻|당🏻🏻|덤🏻|🌺|🍊|🍉|🍑|🍅|🦴|🙌🏻|.🏻|😅|🤔|🍌|🧁|🧡|😐|😞|🍷|🤣|🌿',"",text)
```

```python
for i in range(len(data['review'])):
    data['review'][i] = rm_emo(data['review'][i])
```

```python
# NNG : '보통명사' / NNP : '고유명사' / XR : '어근' / 'VA' : 형용사
data['summary_text'] = '-'
for i in range(len(data['review'])):
    tx = kkma.pos(data['review'][i])
    ttx = pd.DataFrame(tx,columns = ['tx','ps'])
    data['summary_text'][i] = ttx[(ttx['ps'] == 'NNG') | (ttx['ps'] == 'NNP') | (ttx['ps'] == 'XR') |(ttx['ps'] == 'VA') ]['tx'].values
```
